3)what are the main differences between lists and tuples in python?
ans:

A list in Python is an ordered, mutable collection of elements that can hold items of different data types, such as integers, strings, floats, or even other lists. Lists are defined using square brackets [ ], and their elements are separated by commas.
Ordered: Elements have a defined sequence and can be accessed by index.
Mutable: Elements can be added, removed, or modified.
Heterogeneous: Can store different data types in a single list.
Dynamic: Size can change as elements are added or removed.

Key Characteristics of Tuples
Ordered: Elements have a defined sequence and can be accessed by index.
Immutable: Elements cannot be added, removed, or changed after the tuple is created.
Heterogeneous: Can store different data types (e.g., integers, strings, floats).
Hashable: Can be used as keys in dictionaries (if all elements are immutable).

4)Describe how dictionaries store data.

In Python, dictionaries store data as key-value pairs using a highly efficient data structure called a hash table. A dictionary is defined by curly braces {} and stores mappings of unique, immutable keys to values, which can be of any data type.
Internal Working:
Hashing Keys:

When a key-value pair is added to a dictionary, the key is passed through a hashing function that generates a unique integer called a hash value.
This hash value determines the storage location (or "bucket") for the key-value pair in memory.
Storing Values:

The computed hash value maps to a specific slot in the dictionaryâ€™s hash table, where the corresponding value is stored.
Collision Handling:

Hash collisions occur when two keys produce the same hash value.
Python uses techniques such as open addressing or chaining to resolve collisions and ensure efficient data storage.

7)How do tuples ensure data integrity in python?

Tuples in Python are immutable, which means once they are created, their contents cannot be modified (no element can be added, removed, or changed). This immutability plays a crucial role in ensuring data integrity by protecting data from accidental or intentional modification.

Ways Tuples Ensure Data Integrity
Immutability Protects Data Consistency:

Since tuples cannot be changed, they help maintain a consistent state for data throughout the program's execution.
Prevents Unintended Changes:

By design, any attempt to modify a tuple results in a TypeError, ensuring data is safeguarded against accidental modification.
Reliable as Dictionary Keys:

Tuples, being hashable, can be used as keys in dictionaries, unlike mutable types like lists. This makes them suitable for mapping complex keys to values.

8) What is a hash table,and how does it relate to dictionaries in Python?
A hash table is a data structure that stores key-value pairs. It uses a hash function to compute a unique index (called a hash value) for each key, which determines where the corresponding value is stored in memory.
Relationship Between Hash Tables and Dictionaries in Python
Python dictionaries (dict) are built on top of hash tables.
The keys in a dictionary are passed through a hash function, and their corresponding values are stored at specific memory locations.
